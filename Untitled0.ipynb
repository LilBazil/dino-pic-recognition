```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

path = Path('/content/drive/MyDrive/нейронка/dinosaur_dataset')

selected_species = [
    "Tyrannosaurus Rex", "Velociraptor", "Triceratops", "Brachiosaurus",
    "Dilophosaurus", "Parasaurolophus", "Spinosaurus", "Compsognathus",
    "Gallimimus", "Stegosaurus", "Pachycephalosaurus", "Ankylosaurus",
    "Corythosaurus", "Microceratus", "Dimorphodon"
]

for species in selected_species:
    species_path = path / species.replace(" ", "_")
    image_files = list(species_path.glob('*.jpg'))[:3]

    plt.figure(figsize=(10, 3))

    for i, img_path in enumerate(image_files, 1):
        try:
            img = Image.open(img_path)
            plt.subplot(1, 3, i)
            plt.imshow(img)
            plt.axis('off')
            plt.title(species.replace('_', ' '))
        except Exception as e:
            print(f"Не получается открыть изображение {img_path}: {e}")

    plt.show()
```

обучение нейронной сети

(на данный момент макс точность 20% и очень долгое обучение по 2-5 часов.
использую vgg16 для тестов)

```python
#работа с операционной системой
import os
#работа с путями
import glob
#для массивов
import numpy as np
#для обработки и анализа данных
import pandas as pd
#предобработка для изменения масштабирования, поворота и др.
from keras.preprocessing.image import ImageDataGenerator
#импорт модели VGG16, которая используется для классификации изображений, а также функции preprocess_input для предварительной обработки входных данных.
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.applications.vgg19 import VGG19
# для построения последовательных моделей нейронных сетей
from keras.models import Sequential
#слои Dense, Flatten и Dropout для построения архитектуры нейронной сети
from keras.layers import Dense, Flatten, Dropout
#оптимизатор Adam для настройки параметров обучения нейронной сети
from keras.optimizers import Adam

# пути для датасетов
train_data_dir = '/content/drive/MyDrive/нейронка2/data'
test_data_dir  = '/content/drive/MyDrive/нейронка2/test'

#создание генераторов данных для обучения и тестирования с масштабированием значений пикселей
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
test_datagen = ImageDataGenerator(rescale=1./255)

#основные переменнные
img_height = 150
img_width = 150
batch_size = 50
num_classes = 10
epochs = 12

#генератор изображений для обучения
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training')
#генератор изображений для валидации
validation_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation')
#генератор изображений для тестирования
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical')

#загрузка редварительно обученной модели VGG16 без верхних слоев для извлечения признаков
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

#веса предварительно обученной модели замораживаются, чтобы не обновлять их в процессе обучения
for layer in base_model.layers:
    layer.trainable = False

# настройка архитектуры модели
# добавлением слоев: Flatten, Dense, Dropout, Softmax для классификации
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# оптимизация параметров обучения
# оптимизатор, функция потерь и метрика для компиляции модели
model.compile(optimizer=Adam(lr=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# доп аугментация для увеличения разнообразия обучающего набора
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

#обучение модели
model.fit(train_generator, epochs=epochs, validation_data=validation_generator)

#оценка модели на тестовом наборе данных для вычисления потерь и точности
loss, accuracy = model.evaluate(test_generator)
print(f'точность на тестовых данных: {accuracy}')

# Предсказание классов для тестовых изображений
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)

# Вывод статистики
print("результаты предсказаний для тестовых изображений:")
for i in range(len(predicted_classes)):
    print(f" {i+1}: предсказанный - {predicted_classes[i]}   (ожидаемый - {test_generator.classes[i]})")

# Анализ точности предсказаний
correct_predictions = np.sum(predicted_classes == test_generator.classes)
total_predictions = len(predicted_classes)
accuracy_percentage = (correct_predictions / total_predictions) * 100
print(f"общая точность предсказаний: {accuracy_percentage}%")
```
